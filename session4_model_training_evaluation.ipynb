## Learning goals

- Build end-to-end ML pipelines
- Use AutoML/HPO techniques responsibly
- Evaluate models with robust protocols, not single-point metrics
- Analyze sensitivity to perturbations (including `tab-err` style tests)

## End-to-end pipeline view

1. Data ingestion and preprocessing
2. Feature transformation
3. Model fitting
4. Validation and model selection
5. Final test and reporting

## AutoML and HPO

### Typical search strategies

- Grid search
- Random search
- Bayesian optimization

### Practical protocol

- Define search space before seeing test results
- Optimize on train/validation or inner folds
- Retrain selected model on train+validation
- Evaluate once on holdout test set

## Robust evaluation

### Beyond a single score

- Classification: precision, recall, F1, PR curves
- Regression: MAE, MSE, $R^2$
- Stability: cross-validation variance and perturbation sensitivity

### Perturbation/sensitivity checks

- Missing-value perturbations
- Noise injections
- Feature corruption and shift scenarios

## Suggested tab-err integration point

If using `tab-err`, run perturbation benchmarks after model selection and include results in the final report.
